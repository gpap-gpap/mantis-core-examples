```python
%load_ext autoreload
%autoreload 2
from thin_layer_tuning_backend import TLE_thin_layer_fitting_plotting as tle
import matplotlib.pyplot as plt
import numpy as np
```

```python
sl = tle.ThinLayerModels("sleipner")
```

```python
fig, ax = plt.subplots(4, 2, figsize=(12, 12), facecolor="w", dpi=300)

ax[0, 0].set_title(f"Sleipner Tuning at 80% water saturation")
ax[0, 1].set_title(f"Sleipner Tuning at 50% water saturation")
for ax1 in ax:
    for ax2 in ax1:
        ax2.set_facecolor("w")
pt = tle.Plots()
sl.saturation_data(bubble_radius=0.000021, patch_size=0.000008)
sl.normal_incidence_traces(saturation=0.8)
sl.tuning_data(saturation=0.8)
sl.freq_data(saturation=0.8)
for model in ["gassmann", "white", "crm1d"]:
    print(
        f"{model} tuning = {sl.tuning['samples'][np.argmax(np.abs(sl.tuning[model]))]}"
    )
pt.plot_saturation_data(
    data=sl.data_vs_saturation,
    ax=ax[0, 0],
    labels={"gassmann": "Gassmann", "crm1d": "crm1d", "white": "White"},
)
pt.plot_tuning_data(
    data=sl.tuning,
    ax=ax[1, 0],
    labels={"gassmann": "Gassmann", "crm1d": "crm1d", "white": "White"},
)
pt.plot_freq_data(
    data=sl.data_vs_freq,
    ax=ax[2, 0],
    labels={"gassmann": "Gassmann", "crm1d": "crm1d", "white": "White"},
)
pt.plot_att_data(
    data=sl.data_vs_freq,
    ax=ax[3, 0],
    labels={"gassmann": "Gassmann", "crm1d": "crm1d", "white": "White"},
)
sl.saturation_data(bubble_radius=0.000042, patch_size=0.0000047)
sl.normal_incidence_traces(saturation=0.5)
sl.tuning_data(saturation=0.5)
sl.freq_data(saturation=0.5)
for model in ["gassmann", "white", "crm1d"]:
    print(
        f"{model} tuning (50%)= {sl.tuning['samples'][np.argmax(np.abs(sl.tuning[model]))]}"
    )
pt.plot_saturation_data(
    data=sl.data_vs_saturation,
    ax=ax[0, 1],
    labels={"gassmann": "Gassmann", "crm1d": "crm1d", "white": "White"},
)
pt.plot_tuning_data(
    data=sl.tuning,
    ax=ax[1, 1],
    labels={"gassmann": "Gassmann", "crm1d": "crm1d", "white": "White"},
)
pt.plot_freq_data(
    data=sl.data_vs_freq,
    ax=ax[2, 1],
    labels={"gassmann": "Gassmann", "crm1d": "crm1d", "white": "White"},
)
pt.plot_att_data(
    data=sl.data_vs_freq,
    ax=ax[3, 1],
    labels={"gassmann": "Gassmann", "crm1d": "crm1d", "white": "White"},
)
plt.tight_layout()
plt.show()
```
