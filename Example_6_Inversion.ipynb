```python
import numpy as np

# from PIL import Image
import matplotlib.image as mplim
from pathlib import Path


def sorted_directory_listing_with_pathlib_glob(directory):
    path_object = Path(directory)
    items = path_object.glob("Vz-*")
    sorted_items = sorted(items, key=lambda item: item.name)
    return [item.name for item in sorted_items]


dic = "./snapshots/baseline/"
files = sorted_directory_listing_with_pathlib_glob(dic)
# file = f"./ccus_baseline/snapshots/Vz-0009000.bin"
snaps = np.empty(shape=(len(files), 2001, 1001))
for i, file in enumerate(files):
    with open(dic + file, "rb") as f:
        temp = np.fromfile(
            f,
            dtype=np.dtype("float32"),
        )
        try:
            print(file)
            snaps[i] = np.reshape(temp, (2001, 1001))
            mplim.imsave(f"{dic}images/Vz/snapshot_{i}.png", snaps[i].T)
            # img = Image.fromarray(snaps[i])
            # img.save(f"./{i}.png")
        except ValueError:
            print("something wrong with the dimensions")
```

```python
import cv2
import os

image_folder = "images"
video_name = "video.avi"

images = [img for img in os.listdir(image_folder) if img.endswith(".png")]
frame = cv2.imread(os.path.join(image_folder, images[0]))
height, width, layers = frame.shape

video = cv2.VideoWriter(video_name, 0, 1, (width, height))

for image in images:
    video.write(cv2.imread(os.path.join(image_folder, image)))

cv2.destroyAllWindows()
video.release()
```
