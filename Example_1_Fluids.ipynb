As presets, the mantis code includes a database of fluid properties relevant to rock physical calculations for carbon dioxide, methane and hydrogen.

```python
fig, ax = manFPlots.initialise_constant_plot(constant="Temperature")
for fluid in ["CarbonDioxide", "Methane"]:
    for temperature in [20, 50]:
        manFPlots.plot_fluid_by_pressure(ax=ax, fluid_name=fluid, temperature=temperature)
ax[1].legend()    
manFPlots.finalise_constant_plot(fig, ax,constant="Temperature")
```

```python
gas = manRP.fluid.Fluid.from_presets(name="Methane", temperature=30, pressure=10)
water = manRP.fluid.Fluid.from_presets(name="Water", temperature=30, pressure=10)
mix = manRP.fluid.FluidMix(fluid1=water, fluid2=gas)
```

```python
fig, ax = manFPlots.initialise_saturation_plot()
manFPlots.fluid_mix_plot(ax=ax,f=mix)
```
